```
import tensorflow as tf
import nltk
nltk.download('punkt')
from nltk import FreqDist
from nltk.util import ngrams    
import pandas as pd
import os
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt 
import numpy as np
import string
import operator
from flair.models import TextClassifier
classifier = TextClassifier.load('en-sentiment')
```

```
from flair.data import Sentence
```

```
df=pd.read_csv("https://raw.githubusercontent.com/mayank7533/Headline-Analysis/master/data.csv",encoding='utf-8')
```

```
translator = str.maketrans('', '', string.punctuation)

df.title=df.title.apply(lambda x:x.lower())
df.title = df.title.apply(lambda x: x.translate(translator))
df.head()
```

```
df['tokens']=df['title'].apply(lambda x:nltk.word_tokenize(x))
df.head()
tok=df['tokens'].tolist()
tokens=[]
for t in tok:
  for tt in t:
    tokens.append(tt)
  
fdist = nltk.FreqDist(tokens)
items=fdist.items()
items=sorted(items,key=operator.itemgetter(1),reverse=True)
print(items)
```

```
def sent(title):
  sentence=Sentence(title)
  classifier.predict(sentence)
  return sentence.labels
df['sentiment']=df.title.apply(lambda x:sent(x) )
df.head()
```

```
def evalSenti(data):
  if str(data)[1]=='P':
    return 1
  else:
    return 0

df.sentiment=df.sentiment.apply(lambda x:evalSenti(x))
df.head()
```

```
df['year']=df['year'].apply(lambda x:str(x))
df['month']=df['month'].apply(lambda x:str(x))
df['date']=df['year'].apply(lambda x:x+"-")+df['month']
df.date.unique()
dfMonthly=pd.DataFrame()
dfYearly=pd.DataFrame()
dfYearly['year']=df.year.unique()
dfYearly['articles']=dfYearly['year'].apply(lambda x:len(df[df['year']==x]))
dfYearly['sentiment']=dfYearly['year'].apply(lambda x:df[df.year==x]['sentiment'].mean())

dfMonthly['month']=df.month.unique()
dfMonthly['articles']=dfMonthly['month'].apply(lambda x:len(df[df.month==x]))
dfMonthly['sentiment']=dfMonthly['month'].apply(lambda x:df[df.month==x]['sentiment'].mean())
months=["Jan","Feb","March","April","May","June","July","Aug","Sep","Oct","Nov","Dec"]
dfMonthly['month']=dfMonthly['month'].apply(lambda x:months[int(x)-1])

plt.plot(dfYearly['year'],dfYearly['articles'])
plt.xlabel('year')
plt.ylabel('No of articles published')
plt.title("Articles published per year")
plt.show()
plt.plot(dfMonthly['month'],dfMonthly['sentiment'])
plt.xlabel('month')
plt.ylabel('Average sentiment (1=positive 2=negative)')
plt.title("Average monthly Sentiment ")
plt.show()
plt.plot(dfMonthly['month'],dfMonthly['articles'])
plt.xlabel('month')
plt.ylabel('No of articles published')
plt.title("Articles published per month")
plt.show()
plt.plot(dfYearly['year'],dfYearly['sentiment'])
plt.xlabel('year')
plt.ylabel('Average sentiment (1=positive 2=negative)')
plt.title("Average yearly Sentiment")
plt.show()
print(dfYearly.head(),dfMonthly.head())
```
